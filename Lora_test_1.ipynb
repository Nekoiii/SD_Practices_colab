```python
import torch
from diffusers import StableDiffusionPipeline, UniPCMultistepScheduler

pipe = StableDiffusionPipeline.from_pretrained(
        "runwayml/stable-diffusion-v1-5",
        torch_dtype=torch.float16,
    ).to('cuda')

pipe.scheduler = UniPCMultistepScheduler.from_config(pipe.scheduler.config)
```

```python
from PIL import Image

def image_grid(imgs, rows, cols):
    assert len(imgs) == rows * cols

    w, h = imgs[0].size
    grid = Image.new("RGB", size=(cols * w, rows * h))
    grid_w, grid_h = grid.size

    for i, img in enumerate(imgs):
        grid.paste(img, box=(i % cols * w, i // cols * h))
    return grid
```

```python
pipe.unet.load_attn_procs("/content/drive/MyDrive/ML_Practices/Lora_test_1/lora/output/pytorch_lora_weights.bin")
```

```python
generator = [torch.Generator(device="cpu").manual_seed(i+1) for i in range(4)]

images = pipe(prompt=" (((shs))),a boy",
              negative_prompt="",
              generator=generator,
              num_inference_steps=50,
              height=512,
              width=512,
              num_images_per_prompt=4,
              guidance_scale=7.0).images

image_grid(images, 2, 2)
```

```python
%%script false --no-raise-error
#https://zhuanlan.zhihu.com/p/616668335
import os;
import re;
import torch;
from safetensors.torch import save_file;

loraName = "shs"

lora_output_dir = '/content/drive/MyDrive/ML_Practices/Lora_test_1/lora/output'

for root, dirs, files in os.walk(lora_output_dir):
  for dir in dirs:
    ckptIndex = re.search('^checkpoint\-(\d+)$', dir);
 
    if ckptIndex:
      newDict = dict();
      checkpoint = torch.load(os.path.join(lora_output_dir, dir, 'pytorch_lora_weights.bin'));
      for idx, key in enumerate(checkpoint):

        newKey = re.sub('\.processor\.', '_', key);
        newKey = re.sub('mid_block\.', 'mid_block_', newKey);
        newKey = re.sub('_lora.up.', '.lora_up.', newKey);
        newKey = re.sub('_lora.down.', '.lora_down.', newKey);
        newKey = re.sub('\.(\d+)\.', '_\\1_', newKey);
        newKey = re.sub('to_out', 'to_out_0', newKey);
        newKey = 'lora_unet_'+newKey;

        newDict[newKey] = checkpoint[key];

      newLoraName = lora_output_dir + '/' + loraName + '-' + ckptIndex.group(1) + '.safetensors';
      print("Saving " + newLoraName);
      save_file(newDict, newLoraName);
```
